## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
db.list_collection_names()
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene' : 20}
field = {'BusinessName' : 1, 'scores.Hygiene' : 1}
results = establishments.find(query, field)
for result in results:
    print(result)
```

```python
# Use count_documents to display the number of documents in the result
hygiene = establishments.count_documents(query)
print(f"The number is establishments with hygiene scores of 20 is: {hygiene}.")
```

```python
# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
result_hygiene = pd.DataFrame(establishments.find(query))
```

```python
# Display the number of rows in the DataFrame
print("Rows in dataframe: ", len(result_hygiene))
```

```python
# Display the first 10 rows of the DataFrame
result_hygiene.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater 
# than or equal to 4.
query_london = {'LocalAuthorityName': {'$regex':'London'}, 'RatingValue': {'$gte': '4'}}
query_london
```

```python
# Use count_documents to display the number of documents in the result
london = establishments.count_documents(query_london)
print(f"The number is establishments with Local Authority of London and RatingValue greater\
    or equal to 4 is: {london}.")
```

```python
# Display the first document in the results using pprint
pprint(establishments.find_one(query_london))
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(establishments.find(query_london))
london_df
```

```python
# Display the number of rows in the DataFrame
print("Rows in dataframe: ", len(london_df))
```

```python
# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
pprint(establishments.find_one({'BusinessName': 'Penang Flavours'},
                              {'geocode.longitude'},
                              {'geocode.latitude'}))
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.0838400

query_penang = {'RatingValue': '5',
               'geocode.latitude': {'$lte': (latitude + degree_search),
                                            '$gte': (latitude - degree_search)},
               'geocode.longitude': {'$lte': (longitude + degree_search),
                                            '$gte': (longitude - degree_search)}}
sort = [('scores.Hygiene', 1)]  

# Print the results
for bob in establishments.find(query_penang).sort(sort):
    pprint(bob)
```

```python
# Convert result to Pandas DataFrame
penang_df = pd.DataFrame(establishments.find(query_penang))
penang_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
hygiene_pipe = {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
groups = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1}}}

# 3. Sorts the matches from highest to lowest
sorting = {'$sort': {'count': -1}}

pipe = [hygiene_pipe, groups, sorting]

pipe_results = list(establishments.aggregate(pipe))

# Print the number of documents in the result
print(len(pipe_results))
```

```python
# Print the first 10 results
pprint(pipe_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
pipeline_df = pd.DataFrame(establishments.find(pipe_results))
pipeline_df
```

```python
# Display the number of rows in the DataFrame
print("Rows in dataframe: ", len(pipeline_df))
```

```python
# Display the first 10 rows of the DataFrame
pipeline_df.head()
```
